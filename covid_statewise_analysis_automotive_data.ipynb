```python
plt.figure(figsize=(10,10))
plt.plot(df["State"],df["Confirmed"],color="deeppink",alpha=1)
plt.title("Confirmed cases")
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Cases confirmed")
plt.legend()
```

```python
plt.plot(df["State"],df["Confirmed"],color="deeppink",alpha=1)
plt.plot(df["State"],df["Recovered"],color="green",alpha=1)
plt.plot(df["State"],df["Deaths"],color="red",alpha=1)
plt.plot(df["State"],df["Active"],color="yellow",alpha=1)

plt.scatter(df["State"],df["Confirmed"],color="deeppink",alpha=1)
plt.scatter(df["State"],df["Recovered"],color="green",alpha=1)
plt.scatter(df["State"],df["Deaths"],color="red",alpha=1)
plt.scatter(df["State"],df["Active"],color="yellow",alpha=1)
plt.xticks(rotation=90)
```
